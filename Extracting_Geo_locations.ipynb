```python
# extracting data from mongoDB
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# Declare the database
db = client.sports247RecruitRanks
collection = db.sports247RecruitLocation
```

```python
# Verify results:
resultsLocs=[]
results = db.sports247RecruitLocation.find()
for result in results:
    resultsLocs.append(result)
```

```python
locations_geo=[]
for i in resultsLocs:
    if i['lat'] != '':
        locations_geo.append(i)
len(locations_geo)
```

```python
locs =[i['location'] for i in locations_geo]
lats =[i['lat'] for i in locations_geo]
lngs =[i['lng'] for i in locations_geo]
len(locs) == len(lats) == len(lngs)

locations
```

```python
locationsDir = {'location': locs, 'lat': lats, 'lng':lngs}
locationsDf = pd.DataFrame(locationsDir)
locationsDf.head()
```
