```python
test = drafted['STAR'].groupby(drafted['SCHOOL']).value_counts()
df_drafted_stars = test.to_frame()
df_drafted_stars = df_drafted_stars.rename(columns={'STAR':'#drafted'})
df_drafted_stars = df_drafted_stars.reset_index()
df_drafted_stars.head(10)
```

```python
school_stars = recruits['STAR'].groupby(recruits['COLLEGE']).value_counts()
df_school_stars = school_stars.to_frame()
df_school_stars = df_school_stars.rename(columns={'STAR':'#recruited'})
df_school_stars = df_school_stars.reset_index()
df_school_stars.head(15)
```

```python
star_merge = pd.merge(df_drafted_stars, df_school_stars, how='left', left_on=['SCHOOL','STAR'], right_on=['COLLEGE','STAR'])
star_merge['COLLEGE'].fillna('N/R', inplace=True)
star_merge['#recruited'].fillna(0, inplace=True)
star_merge['#recruited'] = star_merge['#recruited'].astype(int)
star_merge
```
