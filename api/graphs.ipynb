```python
import squarify
```

```python
x = 0.
y = 0.
width = 100.
height = 100.

values = [884, 1279, 287, 32]

normed = squarify.normalize_sizes(values, width, height)
rects = squarify.squarify(normed, x, y, width, height)

# Choose colors from http://colorbrewer2.org/ under "Export"
color_brewer = ['rgb(166,206,227)','rgb(31,120,180)','rgb(178,223,138)',
                'rgb(51,160,44)','rgb(251,154,153)','rgb(227,26,28)']
shapes = []
annotations = []
counter = 0

for r in rects:
    shapes.append( 
        dict(
            type = 'rect', 
            x0 = r['x'], 
            y0 = r['y'], 
            x1 = r['x']+r['dx'], 
            y1 = r['y']+r['dy'],
            line = dict( width = 2 ),
            fillcolor = color_brewer[counter]
        ) 
    )
    annotations.append(
        dict(
            x = r['x']+(r['dx']/2),
            y = r['y']+(r['dy']/2),
            text = values[counter],
            showarrow = False
        )
    )
    counter = counter + 1
    if counter >= len(color_brewer):
        counter = 0

# For hover text
trace0 = go.Scatter(
    x = [ r['x']+(r['dx']/2) for r in rects ], 
    y = [ r['y']+(r['dy']/2) for r in rects ],
    text = [ str(v) for v in values ], 
    mode = 'text',
)
        
layout = dict(
    height=700, 
    width=700,
    xaxis=dict(showgrid=False,zeroline=False,showticklabels=False),
    yaxis=dict(showgrid=False,zeroline=False,showticklabels=False),
    shapes=shapes,
    annotations=annotations,
    hovermode='closest'
)

# With hovertext
figure = dict(data=[trace0], layout=layout)

# Without hovertext
# figure = dict(data=[Scatter()], layout=layout)

py.iplot(figure, filename='squarify-treemap-test')
```

```python
five_recruited = recruits.loc[recruits['STAR'] == 5]
five_recruited = five_recruited.groupby('COLLEGE').count().sort_values('PLAYER', ascending=False)
five_recruited = five_recruited['PLAYER'].to_frame().rename(columns={'PLAYER':'RECRUITED'})
```

```python
five_drafted = drafted.loc[drafted['STAR'] == 5]
five_drafted = five_drafted.groupby('COLLEGE_NFL').count().sort_values('PLAYER', ascending=False)
five_drafted = five_drafted['PLAYER'].to_frame()#.rename(columns={'PLAYER':'DRAFTED'}).reset_index()
five_drafted = five_drafted.reset_index().rename(columns={'PLAYER':'DRAFTED','COLLEGE_NFL':'COLLEGE'}).set_index('COLLEGE')
```

```python
draft_ratio = five_drafted.join(five_recruited, how='outer')
draft_ratio = draft_ratio.fillna(0).astype(int)
draft_ratio['RATIO'] = draft_ratio['DRAFTED']/draft_ratio['RECRUITED']
```

```python
ratio2 = draft_ratio.replace([np.inf, -np.inf], np.nan).dropna(how="any")

ratio2 = ratio2.loc[ratio2['RECRUITED'] >= 2]
ratio2 = ratio2.sort_values('RATIO', ascending=False).reset_index().head(20)
ratio2.head()
```

```python
x_val = ratio2['COLLEGE'].tolist()
y_val = ratio2['RATIO'].tolist()
d_marker = ratio2['DRAFTED'].tolist()
r_marker = ratio2['RECRUITED'].tolist()
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

drafted = go.Scatter(
    x=x_val,
    y=y_val,
    text=d_marker,
    mode='markers',
    marker=dict(
        size=d_marker,
        sizemode='area',
        sizeref=2.*max(d_marker)/(40.**2),
        sizemin=4,
        color='blue'
    )
)


data = [drafted]
py.iplot(data, filename='bubblechart-size')
```
