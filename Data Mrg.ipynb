```python
# Dependencies
import pandas as pd
```

```python
recruitsDf = pd.read_csv('01recruitsDf_noGeoLocs.csv', index_col=0)
locationsDf= pd.read_csv('locationsDf.csv', index_col=0)
```

```python
recruitsDf.head()
```

```python
recruitsDf.shape
```

```python
locationsDf.head()
```

```python
locationsDf.shape
```

```python
recruitsDf_WithGeoLocs = pd.merge(recruitsDf, locationsDf, on='location')
recruitsDf_WithGeoLocs.shape
recruitsDf_WithGeoLocs.to_csv('02recruitsDf_WithGeoLocs.csv')
```

```python
recruitsDf_WithGeoLocs
```

```python
import matplotlib.pyplot as plt
```

```python
plot_df = recruitsDf_WithGeoLocs[['lat', 'lng']].groupby(['lat','lng']).size().reset_index(name='count')
plot_df.head()
```

```python
plot_df.plot.scatter(x='lat', y='lng')
plt.show()
```

```python
plot_df.plot.scatter(x='lat', y='lng', s=1 * plot_df['count'],
                     c='count', cmap='viridis')
plt.show()
```
