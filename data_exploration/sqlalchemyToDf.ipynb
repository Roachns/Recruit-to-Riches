```python
# Create an engine for the database 
from sqlalchemy import create_engine
engine = create_engine("sqlite:///football.db") # SQLite DataBase 
type(engine) # QC step
```

```python
# Use the Inspector to explore the database and print the table names
from sqlalchemy import inspect # 
inspector = inspect(engine) # make inpector using the engine 
db_tables=inspector.get_table_names() # from inspector get table names

db_tables # QC Step - check if names consistant with db
```

```python
# Use Inspector to print the column names and types
recruitsCols = inspector.get_columns(db_tables[1])
for c in recruitsCols:
    print(c['name'], c["type"])
```

```python
# list of cols 
recruitsCols=[c['name'] for c in recruitsCols]
recruitsCols
```
